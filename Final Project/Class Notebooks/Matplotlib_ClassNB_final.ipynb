## 1 Histogram : We should use histogram when we need the count of the variable in a plot.

```python
#national park dataset
import pandas as pd
import matplotlib.pyplot as plt
np_data=pd.read_csv("national_parks.csv")
```

```python
np_data.head()
```

```python
np_data.info()
```

```python
plt.hist(np_data['GrandCanyon'],
        facecolor="lightgreen",
        edgecolor="blue",
        bins=10)
plt.xlabel("Count")
plt.ylabel("Frequency")
plt.title("Histogram plot of Grand Canyon")
plt.show()
```

#### The cumulative property
If True, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values. The last bin gives the total number of datapoints.

```python
plt.hist(np_data['GrandCanyon'],
        facecolor="lightgreen",
        edgecolor="blue",
        bins=10,
        cumulative=True)
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram plot of Grand Canyon")
plt.show()
```

#### Check the histogram to a range of values
We only look at the data points within the range 2M-5M. This realigns the bins in the histogram

```python
plt.hist(np_data['GrandCanyon'],
        facecolor="lightgreen",
        edgecolor="blue",
        bins=10,
        range=(2000000, 5000000))
plt.xlabel("Range 2M to 5 M")
plt.ylabel("Frequency")
plt.title("Histogram plot of Grand Canyon")
plt.show()
```

## Multiple Histogram

```python
plt.hist(np_data['GrandCanyon'],
        facecolor="lightgreen",
        edgecolor="blue",
        bins=10)
plt.hist(np_data['BryceCanyon'],
        facecolor="orangered",
        edgecolor="maroon",
        bins=10)
plt.hist(np_data['Badlands'],
        facecolor="blue",
        edgecolor="red",
        bins=3)
plt.legend(["Grand Canyon", "BryceCanyon", "Badlands"], loc=9)
plt.xlabel("Range")
plt.ylabel("Frequency")
plt.title("Histogram plot of Grand Canyon, Bryce Canyon and Badlands")
plt.show()
```

## 2  Pie Chart: Pie chart shows distribution of data based on proportion of pie occupied. 

#### Importing Libraries and Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#movie data
movie_data=pd.read_csv("types_movies.csv")
```

```python
movie_data.head()
```

```python
plt.pie(movie_data['Percentage'],
       labels=movie_data['Sector'])
plt.title("Pie plot of Movie data")
plt.show()
```

#### Pie chart customizations
 We set the values for the colors and autopct properties. The latter sets the format for the values to be displayed. 

https://matplotlib.org/stable/gallery/color/named_colors.html

```python
colors_list=['darkorange','sandybrown','darksalmon','orangered','chocolate']
```

```python
plt.pie(movie_data['Percentage'],
       labels=movie_data['Sector'],
       colors=colors_list,
       autopct="%.0f")
plt.title("Pie plot of Movie data")
plt.show()
```

#### The explode property
To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart. 

The value for "explode" represents the fraction of the radius with which to offset each wedge.

```python
explode_list=[0, .3, 0,0, 0]
plt.pie(movie_data['Percentage'],
       labels=movie_data['Sector'],
       colors=colors_list,
        explode=explode_list,
       autopct="%.0f")
plt.title("Pie plot of Movie data")
plt.show()
```

## 4 Boxplot 
Boxplot gives a nice summary of the data. It helps in understanding our distribution better.

* The end of the box shows the upper quartile(75%)and the start of the box represents the lower quartile(25%). 
* Upper Quartile is also called 3rd quartile and similarly, Lower Quartile is also called as 1st quartile. 
* The region between lower quartile and the upper quartile is called as Inter Quartile Range(IQR) and it is used to approximate the 50% spread in the middle data(75â€“25=50%).

```python
import matplotlib.pyplot as plt
import pandas as pd
# exams
exam_data=pd.read_csv("exams.csv")
```

```python
exam_data.head()
```

```python
exam_data.describe()
```

```python
exam_data.columns
```

```python
exam_scores=exam_data[['math score', 
                       'reading score',
                       'writing score']]
```

```python
exam_scores.head()
```

###  Box plots

```python
colors=['blue', 'red', 'green']
BP=plt.boxplot(exam_scores,
           patch_artist=True,
           notch=True)
plt.xticks([1,2,3],['math score', 'reading score', 'writing score'])
plt.title("Box plot of exam scores")
plt.ylabel("scores")

for patch, color in zip(BP['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```

### Violin Plots
Violin plot is a better chart than boxplot as it gives a much broader understanding of the distribution. It resembles a violin and dense areas point the more distribution of data otherwise hidden by box plots


```python
plt.violinplot(exam_scores,
              showmedians=True)
plt.xticks([1,2,3],['math score', 'reading score', 'writing score'])
plt.title("Violin plot of exam scores")
plt.ylabel("scores")
plt.show()
```

## TwinAxis
* TwinAxis helps in visualizing plotting 2 plots w.r.t to the y-axis and same x-axis.
* It should when we require 2 plots or grouped data in the same direction.

```python
import pandas as pd
import matplotlib.pyplot as plt
# austin weather
austin_weather=pd.read_csv("austin_weather.csv")
austin_weather.head()
```

```python
austin_weather.info()
```

#### Extracting Date,Avg Temperature and Avg Wind Speed columns

```python
data_weather=austin_weather[['Date', 
                             'TempAvgF', 
                             'WindAvgMPH']].head(30)
data_weather.head()
```

```python
data_weather['Date']=pd.to_datetime(data_weather['Date'])
```

### Plotting two lines with 2 separate Y axes with the same X axis
First create one axis and then use the twinx() function to create another axis sharing the same X axis


```python
fig,ax1=plt.subplots()
fig=plt.figure(figsize=(12,6))
#set x label
ax1.set_xlabel("Date")
#add tick parameters , bottom label disable
ax1.tick_params(axis="x",
               bottom=False,
               labelbottom=False)
# set y label in the left side
ax1.set_ylabel("Temp(F)",
              color="red",
              size="x-large")
# tick params of y
ax1.tick_params(axis="y",
               labelcolor="red",
               labelsize="large")
# plot avg Temp and Date
ax1.plot(data_weather['Date'],
        data_weather['TempAvgF'],
        color="red")
ax2=ax1.twinx()
# set y label
ax2.set_ylabel("Avg Wind Speed (mph)",
              color="blue",
              size="x-large")
# ticks
ax2.tick_params(axis="y",
               labelcolor="blue",
               labelsize="large")
ax2.plot(data_weather['Date'],
        data_weather['WindAvgMPH'],
        color="blue")
plt.show()
```

### Twin axis for the same data
If the same data can be expressed in different units, we can configure our axis to display both units on two axes.


```python
def f2c(f):
    return (f-32)*5/9
```

```python
fig,ax1=plt.subplots()
fig=plt.figure(figsize=(12,6))
#set x label
ax1.set_xlabel("Date")
#add tick parameters , bottom label disable
ax1.tick_params(axis="x",
               bottom=False,
               labelbottom=False)
# set y label in the left side
ax1.set_ylabel("Temp(F)",
              color="red",
              size="x-large")
# tick params of y
ax1.tick_params(axis="y",
               labelcolor="red",
               labelsize="large")
# plot avg Temp and Date
ax1.plot(data_weather['Date'],
        data_weather['TempAvgF'],
        color="green")
ax2=ax1.twinx()
# get the minimum and maximum of average temp in farenheit
ymin,ymax=ax1.get_ylim()
#print(ymin)
# set y min and y max for average temp in celsius
ax2.set_ylim(f2c(ymin), f2c(ymax))
ax2.set_ylabel("Temp(C)",
              color="blue",
              size="x-large")
# tick params of y
ax2.tick_params(axis="y",
               labelcolor="blue",
               labelsize="large")

plt.show()
```

## 6 Stack Plot 
* Stack plot visualizes data in stacks and shows the distribution of data over time.
* It is used for checking multiple variable area plots in a single plot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data_np=pd.read_csv("national_parks.csv")
data_np.head()
```

```python
x=data_np['Year']
y=np.vstack([data_np['Badlands'],
           data_np['GrandCanyon'],
             data_np['BryceCanyon']])
```

```python
data_np.columns
```

```python
labels=['Badlands', 'GrandCanyon', 'BryceCanyon']
```

```python
colors=['green', 'red', 'blue']
plt.stackplot(x,y, labels=labels,
             colors=colors,
             edgecolor='black')
plt.legend(loc=2)
plt.show()
```

### Percentage stack plot

```python
data_percent=data_np.divide(data_np.sum(axis=1), axis=0)
```

```python
plt.stackplot(x,data_percent['Badlands'], 
              data_percent['GrandCanyon'],
              data_percent['BryceCanyon'],
              labels=labels,
             colors=colors,
             edgecolor='black')
plt.legend(loc="center left", bbox_to_anchor=(1,0.5))
plt.show()
```

## 7 Bar Plot
* Bar Plot shows the distribution of data over several groups. 
* It is commonly confused with a histogram which only takes numerical data for plotting. 
* It helps in comparing multiple numeric values.

```python
import matplotlib.pyplot as plt
import pandas as pd
countries=pd.read_csv("countries.csv")
countries.head()
```

```python
countries['year'].value_counts()
```

```python
data_2007=countries[countries['year']==2007]
data_2007.info()
```

##### selecting top 10 populated countries

```python
top_10_p=data_2007.sort_values('population', 
                               ascending=False).head(10)
```

```python
top_10_p.head()
```

```python
plt.figure(figsize=(12,6))
x=range(10)
plt.bar(x, top_10_p['population'])
plt.xticks(x,top_10_p['country'])
plt.xlabel("Countries")
plt.ylabel("Population in Bilions")
plt.title("10 most populated countries in 2007")
plt.show()
```

#### Plot mean life expentency of different continents

```python

grouped_data=countries.groupby('continent')['lifeExpectancy'].mean()
print(grouped_data)

x=range(5)
plt.bar(x, grouped_data)
plt.xticks(x, grouped_data.index)
plt.xlabel("Continents")
plt.ylabel("Average life expectancy")
plt.show()
```
